# IT Viec

```python
html = requests.get('https://itviec.com/it-jobs/data-analyst?job_selected=senior-data-analyst-up-to-1800-gross-okxe-3524')
# html.encoding = 'utf-8'
sp = bs(html.text, 'lxml')
print(sp)
```

## get links

```python
tags = sp.find_all('h3', attrs={'class':'title job-details-link-wrapper'})
a_tags = []
jobs = []
links = []
# # # for tag in tags:
for tag in tags:
    a_tags.append(tag.find('a'))
# tags
for a in a_tags:
    links.append(a['href'])
for a in a_tags:
    jobs.append(a.text)
```

```python
import pandas as pd
```

```python
itviec = pd.DataFrame({'web_name':'IT Viec',
                      'jobs': jobs,
                      'links': links})
```

```python
itviec.links = itviec.links.map(lambda x: 'https://itviec.com' + x)
```

```python
itviec.to_csv('itviec.csv')
```

# Create Function to read a link

```python
def read_link(link):
    html = requests.get(link)
    # html.encoding = 'utf-8'
    sp = bs(html.text, 'lxml')
    details = sp.find_all('div', 'job-details__paragraph' )
    divs=[]
    for div in details:
        divs.append(div.find('ul'))
    divs
    dictionary = {}
    for i, ul in enumerate(divs):
        info[i]=ul.text.strip()
    # Convert to dataframe
    df1 = pd.DataFrame(info, index=[0])
    df1.columns= ['job_description', 'requirement', 'benefit']
    company = sp.find('div', "employer-long-overview").find('h3').text
    other = sp.find('div', "employer-long-overview").find_all('div', 'svg-icon__text')
    dict_ = {}
    for i, txt in enumerate(other):
        dict_[i] = txt.text.strip()
    df2 = pd.DataFrame(dict_, index=[0])
#     df2.columns=['Type_of_company', 'scale', 'time','Orginin', 'Other']
    df = pd.concat([df1,df2], axis=1)
    return df
    

    
```
